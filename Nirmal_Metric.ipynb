```python
def addProbAcceptedAnswerToTable(answersTable_provided,questionsTable_provided):
    
    answersTable=answersTable_provided.copy()
    questionsTable=questionsTable_provided.copy()
    answersTable=answersTable.sort_values([' OwnerUserId',' CreationDate']) #order by user and then by creation date
    questionsTable=questionsTable.sort_values([' OwnerUserId',' CreationDate']) #order by user and then by creation date
    
    #Join the answers and questions table
    ansQ = pd.merge(answersTable, questionsTable, how = 'inner', left_on = ' ParentId', right_on = "Id", suffixes=('_ans', '_ques'))
    
   
    #User gets -1 if they pose a question
    questionsTableO=ansQ.sort_values([' OwnerUserId_ques',' CreationDate_ques']) #order by user
    questionsTableO["test"] = ~questionsTableO[["Id_ques", ' OwnerUserId_ques']].duplicated()
    questionsTableO["outdeg"] = np.where(questionsTableO['test']==True, -1,0)


    #users get +1 every time they answer a question of a new user
    answersTableO=ansQ.sort_values([' OwnerUserId_ans',' CreationDate_ans']) #order by user and creation date
    answersTableO["test"] = ~answersTableO[[" OwnerUserId_ans", ' OwnerUserId_ques']].duplicated()
    answersTableO["indeg"] = np.where(answersTableO['test']==True, 1,0)
    
    # some preprocessing to merge the two
    degreeTable1 = questionsTableO[[" OwnerUserId_ques", ' CreationDate_ques', 'outdeg']]
    degreeTable1 = degreeTable1.rename(index = str, columns={" OwnerUserId_ques":"UserId", ' CreationDate_ques': "CreationDate",  'outdeg':"degree" })

    degreeTable2 = answersTableO[[" OwnerUserId_ans", ' CreationDate_ans', 'indeg']]
    degreeTable2 = degreeTable2.rename(index = str, columns={" OwnerUserId_ans":"UserId", ' CreationDate_ans': "CreationDate",  'indeg':"degree" })

    #the degree metric summary table
    degreeTableSum = pd.concat([degreeTable1,degreeTable2])
    degreeTableSum = degreeTableSum.sort_values(['UserId', 'CreationDate'])
    degreeTableSum["cumnetdeg"] = degreeTableSum.groupby('UserId')['degree'].cumsum()
    
    answersTable["cumnetdeg"] = pd.merge(answersTable, degreeTableSum, how = 'inner', left_on = [' OwnerUserId', ' CreationDate'], right_on = ["UserId", "CreationDate"], suffixes=('_ans', '_ques'))["cumnetdeg"]

    return(answersTable)
 
```

```python
RanswersTable = addProbAcceptedAnswerToTable(answersTable,questionsTable)
```

```python
def formatCreationDateColumn(table_provided):
    table = table_provided.copy()
    
   
    #Correcting format of dates in the table
    table[' CreationDate']=pd.to_datetime(table[' CreationDate'],errors="coerce")
    
    #remove lines which could not be formatted to correct date format
    table=table.loc[~table[" CreationDate"].isnull()]
    
    return(table)
```

```python
def gettingTableUserFirstActivity(answersTable_provided,questionsTable_provided):
    answersTable = answersTable_provided.copy()
    questionsTable = questionsTable_provided.copy()
    
    #"Normalization": 0 should correspond to time of first activity so that time is read as days after first activity
    #identify the first activity of users
    answersTable["FirstActivity"]=~answersTable[" OwnerUserId"].duplicated()
    questionsTable["FirstActivity"]=~questionsTable[" OwnerUserId"].duplicated()
    
    #obtaining first answer and first question of a user
    firstAnswers=answersTable.loc[answersTable["FirstActivity"]==True][[" OwnerUserId"," CreationDate"]]
    firstQuestions = questionsTable.loc[questionsTable["FirstActivity"]==True][[" OwnerUserId"," CreationDate"]]
    
    #and merge them into one table
    firstActTable=firstQuestions.merge(firstAnswers, left_on=" OwnerUserId", right_on=" OwnerUserId", how='outer')
    firstActTable.columns = [" OwnerUserId","FirstQuestion","FirstAnswer"]
    
    #obtain first activity (either answer or question)
    firstActTable["FirstActivity"]=firstActTable[["FirstQuestion","FirstAnswer"]].min(axis=1)
    
    return(firstActTable)
```

```python
RfirstActTable = gettingTableUserFirstActivity(RanswersTable,questionsTable)
```

```python
def obtainsummaryTableDegreeUser(answersTable,questionsTable,firstActTable):
    
    #Adding first activity table to answerstable 
    summaryTableDegreeUser = answersTable[["Id"," OwnerUserId"," CreationDate"," ParentId","cumnetdeg"]]
    summaryTableDegreeUser=summaryTableDegreeUser.merge(firstActTable,left_on = " OwnerUserId",right_on=" OwnerUserId")
    
    #Use first activity to obtain normalized time
    summaryTableDegreeUser["Present"]= max(summaryTableDegreeUser[" CreationDate"])
    #timestamp of answer compared to first activity of that user
    summaryTableDegreeUser["DifferenceNumerator"]=(summaryTableDegreeUser[" CreationDate"]-summaryTableDegreeUser["FirstActivity"])
    summaryTableDegreeUser["DifferenceDenominator"]=(summaryTableDegreeUser["Present"]-summaryTableDegreeUser["FirstActivity"])
    
    summaryTableDegreeUser.DifferenceNumerator=summaryTableDegreeUser.DifferenceNumerator.dt.total_seconds() /(24*3600) # to obtain days with decimals
    summaryTableDegreeUser.DifferenceDenominator=summaryTableDegreeUser.DifferenceDenominator.dt.total_seconds()/(24*3600)
    summaryTableDegreeUser["NormTime"]=summaryTableDegreeUser["DifferenceNumerator"]/summaryTableDegreeUser["DifferenceDenominator"]
    
    return(summaryTableDegreeUser)
```
