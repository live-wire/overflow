```python
answersTable=answersTable.sort_values([' OwnerUserId',' CreationDate']) #order by user
questionsTable=questionsTable.sort_values([' OwnerUserId',' CreationDate']) #order by user
```

```python
ansQ=pd.match(answersTable[" ParentId"].tolist(),questionsTable['Id'])
```

```python
ansQ = pd.merge(answersTable, questionsTable, how = 'inner', left_on = ' ParentId', right_on = "Id", suffixes=('_ans', '_ques'))
ansQ
```

```python
users = ansQ.sort_values([' OwnerUserId_ans',' CreationDate_ans', ' OwnerUserId_ques', ' CreationDate_ques'])
users = users[['Id_ans', ' OwnerUserId_ans',' CreationDate_ans', 'Id_ques', ' OwnerUserId_ques', ' CreationDate_ques']]
users
```

```python
#Normalization: 0 should correspond to time of first activity and 1 to present date. 
#1 converting creation date to proper time format
ansQ[' CreationDate_ans']=pd.to_datetime(ansQ[' CreationDate_ans'])
ansQ[' CreationDate_ques']=pd.to_datetime(ansQ[' CreationDate_ques'])
```

```python

questionsTableO=ansQ.sort_values([' OwnerUserId_ques',' CreationDate_ques']) #order by user
```

```python
questionsTableO["test"] = ~questionsTableO[["Id_ques", ' OwnerUserId_ques']].duplicated()
questionsTableO["outdeg"] = np.where(questionsTableO['test']==True, -1,0)
##questionsTableO['outdeg'] = questionsTableO.groupby(' OwnerUserId_ques')["test1"].cumsum()
```

```python
answersTableO=ansQ.sort_values([' OwnerUserId_ans',' CreationDate_ans']) #order by user
```

```python
answersTableO["test"] = ~answersTableO[[" OwnerUserId_ans", ' OwnerUserId_ques']].duplicated()
answersTableO["indeg"] = np.where(answersTableO['test']==True, 1,0)
# answersTableO["test2"] = ~answersTableO[[" OwnerUserId_ans", ' OwnerUserId_ques']].duplicated()
# answersTableO['ques_count'] = answersTableO.groupby(' OwnerUserId_s')["test1"].cumsum()
```

```python
degreeTable1 = questionsTableO[[" OwnerUserId_ques", ' CreationDate_ques', 'outdeg']]
degreeTable1 = degreeTable1.rename(index = str, columns={" OwnerUserId_ques":"UserId", ' CreationDate_ques': "CreationDate",  'outdeg':"degree" })

degreeTable2 = answersTableO[[" OwnerUserId_ans", ' CreationDate_ans', 'indeg']]
degreeTable2 = degreeTable2.rename(index = str, columns={" OwnerUserId_ans":"UserId", ' CreationDate_ans': "CreationDate",  'indeg':"degree" })
```

```python
degreeTableSum = pd.concat([degreeTable1,degreeTable2])
degreeTableSum = degreeTableSum.sort_values(['UserId', 'CreationDate'])
degreeTableSum["cumnetdeg"] = degreeTableSum.groupby('UserId')['degree'].cumsum()
degreeTableSum.loc[degreeTableSum["UserId"]==1968]
```

```python
#Normalization: 0 should correspond to time of first activity and 1 to present date. 
#1 converting creation date to proper time format
answersTable[' CreationDate']=pd.to_datetime(answersTable[' CreationDate'])
questionsTable[' CreationDate']=pd.to_datetime(questionsTable[' CreationDate'])
answersTable["FirstActivity"]=~answersTable[" OwnerUserId"].duplicated()
questionsTable["FirstActivity"]=~questionsTable[" OwnerUserId"].duplicated()
firstAnswers=answersTable.loc[answersTable["FirstActivity"]==True][[" OwnerUserId"," CreationDate"]]
firstQuestions = questionsTable.loc[questionsTable["FirstActivity"]==True][[" OwnerUserId"," CreationDate"]]
firstActTable=firstQuestions.merge(firstAnswers, left_on=" OwnerUserId", right_on=" OwnerUserId", how='outer')
firstActTable.columns = [" OwnerUserId","FirstQuestion","FirstAnswer"]
firstActTable["FirstActivity"]=firstActTable[["FirstQuestion","FirstAnswer"]].min(axis=1)

firstActTable
```

```python
degreeTable=degreeTableSum.merge(firstActTable,left_on = "UserId",right_on=" OwnerUserId")
degreeTable
degreeTable["Present"]= max(degreeTable["CreationDate"])
degreeTable["DifferenceNumerator"]=(degreeTable["CreationDate"]-degreeTable["FirstActivity"])
degreeTable["DifferenceDenominator"]=(degreeTable["Present"]-degreeTable["FirstActivity"])
degreeTable.DifferenceNumerator=degreeTable.DifferenceNumerator.dt.days
degreeTable.DifferenceDenominator=degreeTable.DifferenceDenominator.dt.days

# degreeTable.DifferenceNumerator=degreeTable.DifferenceNumerator.dt.days
# degreeTable.DifferenceDenominator=degreeTable.DifferenceDenominator.dt.days
#degreeTable["NormTime"]=degreeTable["DifferenceNumerator"]/degreeTable["DifferenceDenominator"]
degreeTable
```

```python
user = 1968 #3732271
plt.figure()
plt.plot(degreeTable.loc[degreeTable["UserId"]==user]["DifferenceNumerator"].tolist(),degreeTable.loc[degreeTable["UserId"]==user]["cumnetdeg"].tolist())
plt.show()
```
